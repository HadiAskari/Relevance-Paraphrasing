```python
accuracy = accuracy_score(truth, prediction)
accuracy
```

```python
precision = precision_score(truth, prediction, average='weighted') #, average='weighted')
precision
```

```python
recall = precision_score(truth, prediction, average='weighted')
recall
```

```python
f1score = f1_score(truth, prediction, average='weighted')
f1score
```

```python
cm=confusion_matrix(truth, prediction, labels=['Left', 'Right'], normalize='all')
```

```python
disp = ConfusionMatrixDisplay(confusion_matrix=cm,
                              display_labels=['Left', 'Right'])
disp.plot()

plt.savefig('results/yt_titles/discrete/cm-llama_2_class_anna.png')
plt.show()
```
